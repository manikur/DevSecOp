```python
P=np.arange(0.,10,0.1)
a=np.cos(P)
b=np.sin(P)
plt.plot(P,a,'c')
plt.plot(P,b,'y')
```

```python
#linewidth
P=np.arange(0.,10,0.1)
a=np.cos(P)
b=np.sin(P)
plt.plot(P,a,linestyle='dotted')
plt.plot(P,b,'y',linewidth=4)
```

```python
#linewidth
P=np.arange(0.,10,0.1)
a=np.cos(P)
b=np.sin(P)
plt.plot(P,a,linestyle='dotted')
plt.plot(P,b,'y',linewidth=4,marker='>',markersize=12,markeredgecolor='red')
```
